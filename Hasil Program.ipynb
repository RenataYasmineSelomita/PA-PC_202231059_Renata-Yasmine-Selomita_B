# IMPORT LIBRARY

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# MEMBACA GAMBAR

```python
img = cv2.imread("Renata_Citra.jpg")
```

# MEMBUAT BARIS DAN KOLOM

```python
img.shape
```

```python
[baris, kolom] = img.shape[:2]
```

# MENGUBAH JADI BGR JADI RGB 

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# MENGUBAH GAMBAR KE GRAYSCALE

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
copyCitraGray = img_gray.copy().astype(float)
```

# DIMENSI GAMBAR

```python
m1, n1 = copyCitraGray.shape
output1 = np.empty([m1,n1])
```

# MANUAL MEAN FILTERING

```python
for baris in range (0, m1-1):
    for kolom in range (0, n1-1):
        a1 = baris
        b1 = kolom
        jumlah = copyCitraGray[a1-1, b1-1] + copyCitraGray[a1-1, b1] + copyCitraGray[a1-1, b1+1] +\
            copyCitraGray[a1, b1-1] + copyCitraGray[a1, b1] + copyCitraGray[a1, b1+1] +\
            copyCitraGray[a1+1, b1-1] + copyCitraGray[a1+1, b1] + copyCitraGray[a1+1, b1+1]
        output1[a1,b1] = 1/9*jumlah
```

# MEDIAN FILTERING MENGGUNAKAN OPENCV

```python
img_median = cv2.medianBlur(img, 5)
```

```python
fig, axis = plt.subplots(2, 2, figsize=(15, 15))
ax = axis.ravel()

ax[0].imshow(img, cmap='gray')
ax[0].set_title('Citra Asli')

ax[1].imshow(img_median, cmap='gray')
ax[1].set_title('After Filter Median')

ax[2].imshow(img_gray, cmap='gray')
ax[2].set_title('Original Image')

ax[3].imshow(output1, cmap='gray')
ax[3].set_title('Mean Filtered Image')

plt.show()
```
